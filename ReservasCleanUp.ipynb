```python
path = os.getcwd()
def limpiarDataset():
    # Initialize an empty list to store DataFrames instead of creating a DataFrame with columns
    dfs = []
    
    fileName = 'Reservas_Diario_20'
    for i in range(4, 22):
        # Create year string
        anyo = f'{i:02d}'  # This handles both single and double-digit years
        
        fullName = fileName + anyo + ".xlsx"
        print("Limpiando:", fullName)
        
        # Read Excel file
        df = pd.read_excel(path + '/' + fullName)
        
        # Clean columns
        df.columns = df.iloc[2]
        #df = df.drop([df.columns[0]], axis=1)
        df = df.drop([0, 1, 2], axis=0)
        
        # Reset index and clean column names
        df = df.reset_index(drop=True)
        df = df.rename(columns={
            'Region Hidrologica': 'RegionHidrologica', 
            'Nombre Embalse': 'NombreEmbalse',
            'Volumen Útil Diario Mm3': 'VolumenUtilDiarioMm3',
            'Volumen Útil Diario Energía kWh': 'VolumenUtilDiarioEnergiaKwh',
            'Volumen Útil Díario %': 'VolumenUtilDiarioPerc',
            'Volumen Mm3':'VolumenMm3',
            'Volumen Energía kWh':'VolumenEnergiaKwh',
            'Volumen %':'VolumenPerc'
            
        })
        
        # Convert data types and handle missing values
        df['Fecha'] = pd.to_datetime(df['Fecha'])
        
        # Fill missing values with mean
        numeric_columns = ['VolumenUtilDiarioMm3', 'VolumenUtilDiarioEnergiaKwh', 'VolumenUtilDiarioPerc','VolumenMm3','VolumenEnergiaKwh','VolumenPerc']
        for col in numeric_columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            df[col] = df[col].fillna(df[col].mean())
        
        # Ensure correct column types
        df = df.astype({
            'RegionHidrologica': str,
            'NombreEmbalse': str,
            'VolumenUtilDiarioMm3': float, 
            'VolumenUtilDiarioEnergiaKwh': float, 
            'VolumenUtilDiarioPerc': float,
            'VolumenMm3': float,
            'VolumenEnergiaKwh': float,
            'VolumenPerc': float
        })
        
        # Append to list of DataFrames
        dfs.append(df)
    
    # Concatenate all DataFrames at once
    dfFinal = pd.concat(dfs, ignore_index=True)
    
    # Save to CSV
    dfFinal.to_csv('dfFinalReservas.csv', encoding='utf-8', index=False)
    
    return dfFinal

limpiarDataset()
```
